```python
# Iris Dataset exercises for Data Mining
# --------------------------------------
# Tashfeen Mustafa Choudhury

# load numpy
import numpy as np

# load matplotlib
import matplotlib
import matplotlib.pyplot as plt

# for normal distribution
import scipy.stats

# load the dataset
from sklearn.datasets import load_iris

iris = load_iris()
```

```python

###
### Extracting Necessary Data

# only load the first two dimensions (sepal length and sepal width)
iris_data = iris.data[:, :2]

# loading the targets
iris_targets = iris.target
print(iris_targets)
'''
0 = setosa
1 = versicolor
2 = virginica
'''

# taking the sepal length feature
x = iris_data[:, 0]

# taking the sepal width feature
y = iris_data[:, 1]

# creating colormap
'''
red = setosa (0)
blue = versicolor (1)
black = virginica (2)
'''
colormap = np.array(['r', 'b', 'g'])

# taking minimum and maximum of sepal length
x_min, x_max = x.min(), x.max()

# taking minimum and maximum of sepal width
y_min, y_max = y.min(), y.max()
```

```python
###
### Answer to Question 6

print('''
6. Draw the normal distribution function for all the sepal length (X1). 
To do that, first calculate sample mean(µ1) and sample variance.
''')

# create figure
plt.figure(7, figsize = (13, 9))
plt.clf()

# Calculate sample sepal length mean (mU)
sepal_length_mean_value = np.mean(x)

# Calculating the sample variance
sepal_length_variance = np.var(x)
print(x)
x_axis = np.linspace(4, 8)
print(x_axis)

x_normal_distribution = scipy.stats.norm.pdf(x_axis, sepal_length_mean_value, sepal_length_variance)
print(x_normal_distribution)

plt.plot(x_axis, x_normal_distribution, color = 'coral')

plt.grid()

plt.xlim(3, 9)
plt.ylim(x_normal_distribution.min() - 0.01, x_normal_distribution.max() + 0.1)

plt.xlabel('Sepal Length')
plt.ylabel('Normal Distribution f(x)')

plt.show()
```

```python
###
### Answer to Question 7

print('''
7. Draw the bivariate normal distribution function for the sepal length (X1) 
and the Sepal Width (X2). So you need to draw the function f(X1,X2). 
To do that, first calculate sample sepal length mean(µ1), 
sample sepal width mean(µ2), and the covariance matrix ∑.
''')

# we already know sepal length mean value and variance of sepal length sample
print('sepal_length_mean_value: ', sepal_length_mean_value)
print('sepal_length_variance: ', sepal_length_variance)

# Calculating sepal width mean value
sepal_width_mean_value = np.mean(y)
print('sepal_width_mean_value: ', sepal_width_mean_value)

# Calculating variance of sepal width
sepal_width_variance = np.var(y)
print('sepal_width_variance: ', sepal_width_variance)

# sigma 12
sigma_12 = (np.sum(np.multiply(xZ.transpose(), yZ)) / 150)
print('sigma_12: ', sigma_12)

# sigma 21
sigma_21 = (np.sum(np.multiply(yZ.transpose(), xZ)) / 150)
print('sigma_21: ', sigma_21)


# Finding 2D Mean
u = np.array([sepal_length_mean_value, sepal_width_mean_value])
uT = u.transpose()
print('uT: ', uT)

# Finding covariance matrix
covariance_matrix = np.cov(x, y)
print('covariance_matrix: ', covariance_matrix)

# correlation
#correlation = covariance_matrix[0][1]/(covariance_matrix[0][0] * covariance_matrix[1][1]) ** 0.5
#print(correlation)

# creating figure 8
plt.figure(8, figsize = (13, 9))
plt.clf()

# create meshgrid, x = sepal length, y = sepal width
X, Y = np.meshgrid(x, y)

# create dstack for .pdf method
dstack = np.dstack((X, Y))

# create multivariate normal
mv_normal = scipy.stats.multivariate_normal(uT, covariance_matrix)

# f(x, y)
Z = mv_normal.pdf(dstack)

# plot 3d axes
ax = plt.axes(projection='3d')

# define the surface
ax.plot_surface(X, Y, Z,cmap='viridis')

plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.show()
```

```python
###
### Answer to Question 1

import pandas as pd

print('''\n
1. Compute the multivariate mean vector
''')

'''
    Column 1.  fLength:  continuous  # major axis of ellipse [mm]
    Column 2.  fWidth:   continuous  # minor axis of ellipse [mm] 
    Column 3.  fSize:    continuous  # 10-log of sum of content of all pixels [in #phot]
    Column 4.  fConc:    continuous  # ratio of sum of two highest pixels over fSize  [ratio]
    Column 5.  fConc1:   continuous  # ratio of highest pixel over fSize  [ratio]
    Column 6.  fAsym:    continuous  # distance from highest pixel to center, projected onto major axis [mm]
    Column 7.  fM3Long:  continuous  # 3rd root of third moment along major axis  [mm] 
    Column 8.  fM3Trans: continuous  # 3rd root of third moment along minor axis  [mm]
    Column 9.  fAlpha:   continuous  # angle of major axis with vector to origin [deg]
    Column 10.  fDist:    continuous  # distance from origin to center of ellipse [mm]
    Column 11.  class:    g,h         # gamma (signal), hadron (background)
'''

# creating list of feature names
featureNames = ['fLength', 'fWidth', 'fSize', 'fConc', 'fConc1', 'fAsym', 
             'fM3Long', 'fM3Trans', 'fAlpha', 'fDist','identity']

# loading magic04 data
magic04_data = pd.read_csv('magic04.data', names=featureNames, header=None)

# converting DataFrame to numpy array
magic04_data_array = magic04_data.to_numpy()

for i in range(len(magic04_data_array)):
  for j in range(len(magic04_data_array[i])):
    if j != len(featureNames) - 1:
      magic04_data_array[i][j] = np.float64(magic04_data_array[i][j])

# fLength: continuous, major axis of ellipse [mm]
fLength = magic04_data_array[:, 0]
# calculating mean for fLength
fLength_mean = np.mean(fLength)

# fWidth: continuous, minor axis of ellipse [mm] 
fWidth = magic04_data_array[:, 1]
# calculating mean for fWidth
fWidth_mean = np.mean(fWidth)

# fSize: continuous, 10-log of sum of content of all pixels [in #phot]
fSize = magic04_data_array[:, 2]
# calculating mean for fSize
fSize_mean = np.mean(fSize)

# fConc: continuous, ratio of sum of two highest pixels over fSize  [ratio]
fConc = magic04_data_array[:, 3]
# calculating mean for fConc
fConc_mean = np.mean(fConc)

# fConc1: continuous, ratio of highest pixel over fSize  [ratio]
fConc1 = magic04_data_array[:, 4]
# calculating mean for fConc1
fConc1_mean = np.mean(fConc1)

# fAsym: continuous, distance from highest pixel to center, projected onto major axis [mm]
fAsym = magic04_data_array[:, 5]
# calculating mean for fAsym
fAsym_mean = np.mean(fAsym)

# fM3Long: continuous, 3rd root of third moment along major axis  [mm]
fM3Long = magic04_data_array[:, 6]
# calculating mean for fM3Long
fM3Long_mean = np.mean(fM3Long)

# fM3Trans: continuous, 3rd root of third moment along minor axis  [mm]
fM3Trans = magic04_data_array[:, 7]
# calculating mean for fM3Trans
fM3Trans_mean = np.mean(fM3Trans)

# fAlpha: continuous, angle of major axis with vector to origin [deg]
fAlpha = magic04_data_array[:, 8]
# calculating mean for fAlpha
fAlpha_mean = np.mean(fAlpha)

# fDist: continuous, distance from origin to center of ellipse [mm]
fDist = magic04_data_array[:, 9]
# calculating mean for fDist
fDist_mean = np.mean(fDist)

# class: g,h gamma (signal), hadron (background)
fClass = magic04_data_array[:, 10]

# creating multivariate mean vector
multivariate_mean_vector = np.array([
  fLength_mean,
  fWidth_mean,
  fSize_mean,
  fConc_mean,
  fConc1_mean,
  fAsym_mean,
  fM3Long_mean,
  fM3Trans_mean,
  fAlpha_mean,
  fDist_mean
])

print('Multivariate Mean Vector: ', multivariate_mean_vector)
```

```python
###
### Answer to Question 2

print('''
2. Compute the sample covariance matrix as inner products between the columns of the centered data matrix
''')
centered_data_matrix = []

for i in range(len(magic04_data_array)):
  row = magic04_data_array[i][0:10]
  centered_data_matrix.append(np.subtract(row, multivariate_mean_vector.transpose()))

centered_data_matrix = np.array(centered_data_matrix)

inner_product = np.dot(centered_data_matrix.transpose(), centered_data_matrix)

# calculating sample covariance matrix
covariance_matrix_using_inner_product = np.multiply(1/19020, inner_product)
print('Sample Covariance Matrix: ', covariance_matrix_using_inner_product)
```

```python
###
### Answer to Question 3

print('''
3. Compute the sample covariance matrix as outer product between the centered data points (see Eq. (2.39) in chapter 2)
''')

sum_of_outer_products = np.zeros((10, 10))

transposed_centered_data_matrix = centered_data_matrix.transpose()

for i in range(len(sum_of_outer_products)):
  for j in range(len(sum_of_outer_products)):
    temp = np.dot(transposed_centered_data_matrix[i], centered_data_matrix[:, j])
    sum = np.sum(temp)
    sum_of_outer_products[i][j] = sum

covariance_matrix_using_outer_product = np.multiply(1/19020, sum_of_outer_products)

print('covariance matrix using outer product: ', covariance_matrix_using_outer_product)
print(covariance_matrix_using_outer_product.shape)
```

```python
print('''
4. Compute the correlation between Attributes 1 and 2 
by computing the cosine of the angle between the centered attribute vectors. 
Plot the scatter plot between these two attributes.
''')

centered_fLength = np.subtract(fLength, fLength_mean)
centered_fWidth = np.subtract(fWidth, fWidth_mean)

inner_product_fLength_fWidth = np.dot(centered_fLength, centered_fWidth)

centered_fLength_norm = np.linalg.norm(centered_fLength)
centered_fWidth_norm = np.linalg.norm(centered_fLength)

correlation_between_fLength_fWidth = (inner_product_fLength_fWidth / centered_fLength_norm * centered_fWidth_norm)

print(correlation_between_fLength_fWidth)

plt.figure(9, figsize = (13, 9))
plt.clf()

plt.scatter(x = centered_fLength, y = centered_fWidth)
plt.grid()
plt.show()
```

```python
print('''
6. Which attribute has the largest variance, and which attribute has the smallest variance? Print these values.
''')

min_variance = 0
max_variance = 0
index_min = 0
index_max = 0

for i in range(len(covariance_matrix_using_outer_product)):
  if i == 0:
    min_variance = covariance_matrix_using_outer_product[i][i] 
    max_variance = covariance_matrix_using_outer_product[i][i] 
  else: 
    if covariance_matrix_using_outer_product[i][i] > max_variance:
      max_variance = covariance_matrix_using_outer_product[i][i]
      index_max = i
    elif covariance_matrix_using_outer_product[i][i] < min_variance:
      min_variance = covariance_matrix_using_outer_product[i][i]
      index_min = i

print('Smallest variance: ', min_variance)
print('Attribute Name: ', featureNames[index_min])
print('Largest variance: ', max_variance)
print('Attribute Name: ', featureNames[index_max])
```

```python
print('''
7.	Which pair of attributes has the largest covariance, and which pair of attributes has the smallest covariance? Print these values.
''')

min_variance = 0
max_variance = 0
index_max_i = 0
index_max_j = 0
index_min_i = 0
index_min_j = 0

for i in range(len(covariance_matrix_using_outer_product)):
  for j in range(len(covariance_matrix_using_outer_product[i])):
    if i != j:
      if min_variance == 0 or max_variance == 0:
        min_variance = covariance_matrix_using_outer_product[i][j]
        max_variance = covariance_matrix_using_outer_product[i][j]
      else:
        if covariance_matrix_using_outer_product[i][i] > max_variance:
          max_variance = covariance_matrix_using_outer_product[i][j]
          index_max_i = i
          index_max_j = j
        elif covariance_matrix_using_outer_product[i][i] < min_variance:
          min_variance = covariance_matrix_using_outer_product[i][j]
          index_min_i = i
          index_min_j = j

print('Smallest covariance: ', min_variance)
print('Attributes: ', featureNames[index_max_i], featureNames[index_max_j])
print('Largest covariance: ', max_variance)
print('Attributes: ', featureNames[index_min_i], featureNames[index_min_j])
```
